```python
import pickle
from igraph import *
import pylab as plt
import numpy as np
import random
import sys
import time
from statsmodels.tools.tools import ECDF
from scipy.stats import gaussian_kde
import math
import statsmodels.api as sm
#from sklearn.naive_bayes import GaussianNB
from scipy.stats import multivariate_normal,gamma
from collections import Counter
from scipy.linalg import solve, sqrtm, norm
%matplotlib inline
```

```python
class Stopwatch:
    start_time=None
    def go(self,msg=''):
        if msg:
            print msg,
        self.start_time=time.time()
        sys.stdout.flush()
    def stop(self,msg=''):
        if msg:
            print "{}: {} seconds".format(msg,time.time()-self.start_time)
        else:
            print "Elapsed time: {} seconds".format(time.time()-self.start_time)
        sys.stdout.flush()
    def check(self):
        return time.time()-self.start_time
```

```python
from IPython import display
fig=plt.figure(figsize=(20,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

tic=Stopwatch()
tic.go()
N=30 # no. of nodes
M=200 # no. of edges
#G=Graph.Erdos_Renyi(n=N,m=M)
#G=Graph.Barabasi(n=N,m=4)
G=Graph.Watts_Strogatz(1, N, 2, 0.1, loops=False, multiple=False)
#G=pickle.load(open('gene_network_0.08.pkl'))
#G=G.subgraph(random.sample(xrange(G.vcount()),N))
#G.simplify()
V=range(G.vcount()) # set of nodes
M=G.ecount()
N=G.vcount()
tic.stop('Load network')

tic.go()
tests=0
Gh=Graph(N,directed=False) # start from an empty graph
tested=dict()
found=0
cost_ls=[0]
testu=Counter()
comp=[]
ncomp=0
compid=dict()
u=random.choice(V)
v=random.choice(V)
while u==v:
    u=random.choice(V)
    v=random.choice(V)
testu.update([u,v])
Z=[u,v]
eid=G.get_eid(u,v,False,False)
tests+=1
if eid==-1:
    tested[tuple(sorted((u,v)))]=0
    comp.append([])
    compid[u]=ncomp
    ncomp+=1
    comp[-1].append(u)
    comp.append([])
    compid[v]=ncomp
    ncomp+=1
    comp[-1].append(v)
else:
    tested[tuple(sorted((u,v)))]=1
    found+=1
    cost_ls.append(tests)
    Gh.add_edge(u,v)
    comp.append([])
    compid[u]=ncomp
    compid[v]=ncomp
    ncomp+=1
    comp[-1]+=[u,v]
    
X0=np.random.randn(2,N)
X=X0.copy()
X2=np.zeros_like(X)
a=2
D=dict()
D[tuple(sorted((u,v)))]=norm(X[:,u]-X[:,v])
iters=500

ax1.clear()
ax1.plot(X[0,Z],X[1,Z],'r.')
display.clear_output(wait=True)
display.display(plt.gcf())
plt.pause(1)
while found<G.ecount():
    Dh=dict()
    for i in Z:
        for j in Z:
            if (j>i) and ((i,j) not in tested):
                Dh[(i,j)]=0
    for t in xrange(iters):
        # work on each cluster first
        for c,C in enumerate(comp):
            if len(C)<=1:
                continue
            D2=dict() # distance between new positions
            for t2 in xrange(1):
                f=1 # likelihood of old positions
                f2=1 # likelihood of new positions
                # update position Z first
                X2[:,C]=X[:,C]+np.random.randn(2,len(C))*0.2 # propose new positions of nodes
                for i in C:
                    for j in C:
                        if (j>i):
                            if (i,j) not in tested:
                                continue
                            D2[(i,j)]=norm(X2[:,i]-X2[:,j])
                            if tested[(i,j)]==1:
                                f2*=1/(1+np.exp(D2[(i,j)]-a))
                                f*=1/(1+np.exp(D[(i,j)]-a))
                            else:
                                f2*=1/(1+np.exp(a-D2[(i,j)]))
                                f*=1/(1+np.exp(a-D[(i,j)]))
                prior=1
                prior2=1
                for i in C:
                    prior*=multivariate_normal.pdf(X[:,i], mean=np.zeros(2), cov=20)
                    prior2*=multivariate_normal.pdf(X2[:,i], mean=np.zeros(2), cov=20)
                if random.random()<f2/f*prior2/prior:
                    D.update(D2)
                    A=X2[:,C]-X2[:,C].mean(axis=1).reshape(2,1)+X0[:,C].mean(axis=1).reshape(2,1)
                    A0=X0[:,C]
                    X[:,C]=A0.dot(A.T).dot(solve(sqrtm(A.dot(A0.T).dot(A0).dot(A.T)),A))
                    #print 'Accept: {}'.format(f2/f*prior2/prior)
                #ax1.plot(X[0,C],X[1,C],'y.')
                #display.clear_output(wait=True)
                #display.display(plt.gcf())
                #plt.pause(1)
        # move each cluster as a whole
        D2=dict()
        f=1
        f2=1
        for C in comp:
            X2[:,C]=X[:,C]+np.random.randn(2,1)
        for i in Z:
            for j in Z:
                if (j>i):
                    if ((i,j) not in tested) or (compid[i]==compid[j]):
                        continue
                    D2[(i,j)]=norm(X2[:,i]-X2[:,j])
                    if tested[(i,j)]==1:
                        f2*=1/(1+np.exp(D2[(i,j)]-a))
                        f*=1/(1+np.exp(D[(i,j)]-a))
                    else:
                        f2*=1/(1+np.exp(a-D2[(i,j)]))
                        f*=1/(1+np.exp(a-D[(i,j)]))
        prior=1
        prior2=1
        for i in Z:
            prior*=multivariate_normal.pdf(X[:,i], mean=np.zeros(2), cov=20)
            prior2*=multivariate_normal.pdf(X2[:,i], mean=np.zeros(2), cov=20)
        if random.random()<f2/f*prior2/prior:
            D.update(D2)
            A=X2[:,Z]-X2[:,Z].mean(axis=1).reshape(2,1)+X0[:,Z].mean(axis=1).reshape(2,1)
            A0=X0[:,Z]
            X[:,Z]=A0.dot(A.T).dot(solve(sqrtm(A.dot(A0.T).dot(A0).dot(A.T)),A))
        # update a
        a2=abs(a+random.uniform(-0.5,0.5))
        f2=1 # likelihood of the new model
        f=1
        for e in tested:
            if tested[e]==1:
                f2*=1/(1+np.exp(D[e]-a2))
                f*=1/(1+np.exp(D[e]-a))
            else:
                f2*=1/(1+np.exp(a2-D[e]))
                f*=1/(1+np.exp(a-D[e]))
        if random.random()<f2/f*gamma.pdf(a2, 1,scale=1)/gamma.pdf(a, 1,scale=1):
            a=a2
        # update distance for untested pairs
        for i in Z:
            for j in Z:
                if (j>i) and ((i,j) not in tested):
                    Dh[(i,j)]+=norm(X[:,i]-X[:,j])
    for e in Dh:
        Dh[e]=1/(1+np.exp(Dh[e]/iters-a))
    if Dh:
        u,v=max(Dh, key=Dh.get)  # test this pair or
    newcomer=0
    if (len(Dh)==0 or random.random()>Dh[(u,v)]) and len(Z)<N:  # add in a new node
        newcomer=1
        u=random.choice(testu.most_common())[0]  # the one with most information
        v=random.choice(np.setdiff1d(V,Z))
        Z.append(v)
        ax1.clear()
        ax1.set_title(str(len(Z)))
    else:
        ax1.clear()
        ax1.plot(X[0,[u,v]],X[1,[u,v]],'bo')
        ax1.set_title(str(Dh[u,v]))
    eid=G.get_eid(u,v,False,False)
    testu.update([u,v])
    tests+=1
    if eid==-1:
        tested[tuple(sorted((u,v)))]=0
        if newcomer==1:
            X[:,v]=np.random.randn(2)*10
            for i in Z:
                D[tuple(sorted([i,v]))]=norm(X[:,i]-X[:,v])
            comp.append([])
            comp[-1].append(v)
            compid[v]=ncomp
            ncomp+=1
    else:
        tested[tuple(sorted((u,v)))]=1
        found+=1
        cost_ls.append(tests)
        Gh.add_edge(u,v)
        if newcomer==1:
            X[:,v]=X[:,u]+np.random.randn(2)*0.2
            for i in Z:
                D[tuple(sorted([i,v]))]=norm(X[:,i]-X[:,v])
            compid[v]=compid[u]
            comp[compid[u]].append(v)
        elif compid[u]!=compid[v]:
            c=compid[v]
            comp[compid[u]]+=comp[c]
            for i in comp[c]:
                compid[i]=compid[u]
            comp[c]=[]
            
    if tests%1==0:
        #ax1.clear()
        ax1.plot(X[0,Z],X[1,Z],'r.')
        for e in tested:
            if tested[e]==1:
                ax1.plot(X[0,e],X[1,e],'k')
        #ax1.set_title(str(len(Z)))
        ax2.plot(tests,a,'b.')
        ax2.set_title(str(tests))
        ax3.clear()
        ax3.plot(np.arange(0.,len(cost_ls))/M, np.array(cost_ls)/(0.5*N*(N-1)))
        ax3.plot([0,1],[0,1],'--',color='grey')
        ax3.set_title("Performance")
        ax3.set_xlabel("fraction of edges discovered")
        ax3.set_ylabel("fraction of pairs tested")
        display.clear_output(wait=True)
        display.display(plt.gcf())
        plt.pause(0.01)
```
